```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 讀取 CSV 檔案
file_path = './Data/CH467160_DiffTurns_AC.csv'  # 確保檔案路徑正確
df = pd.read_csv(file_path)

# 整理數據
columns = [
    "Bm_N4", "Pcv_N4", "Bm_N10", "Pcv_N10", "Bm_N20", "Pcv_N20", "Bm_N36",
    "Pcv_N36"
]
df_cleaned = df.iloc[1:].reset_index(drop=True)  # 移除第一行標題
df_cleaned.columns = columns  # 重新命名欄位
df_cleaned = df_cleaned.astype(float)  # 轉換數據類型
```

```python
# ======= 線性尺度圖 =======
plt.figure(figsize=(8, 6))
plt.plot(df_cleaned["Bm_N4"],
         df_cleaned["Pcv_N4"],
         'o--',
         label="N=4",
         linewidth=2,
         markersize=6)
plt.plot(df_cleaned["Bm_N10"],
         df_cleaned["Pcv_N10"],
         's--',
         label="N=10",
         linewidth=2,
         markersize=6)
plt.plot(df_cleaned["Bm_N20"],
         df_cleaned["Pcv_N20"],
         '^--',
         label="N=20",
         linewidth=2,
         markersize=6)
plt.plot(df_cleaned["Bm_N36"],
         df_cleaned["Pcv_N36"],
         'd--',
         label="N=36",
         linewidth=2,
         markersize=6)

# 標籤與標題
plt.xlabel(r"$B_{m}$ (mT)", fontsize=12)
plt.ylabel(r"$P_{core}$ (kW/m³)", fontsize=12)
plt.title("Different Turns Compare (Linear Scale)", fontsize=14)
plt.legend(fontsize=10)
plt.grid(which="both", linestyle="--", linewidth=0.5)

# 顯示圖表
plt.show()

# ======= 對數尺度圖 =======
plt.figure(figsize=(8, 6))
plt.plot(df_cleaned["Bm_N4"],
         df_cleaned["Pcv_N4"],
         'o--',
         label="N=4",
         linewidth=2,
         markersize=6)
plt.plot(df_cleaned["Bm_N10"],
         df_cleaned["Pcv_N10"],
         's--',
         label="N=10",
         linewidth=2,
         markersize=6)
plt.plot(df_cleaned["Bm_N20"],
         df_cleaned["Pcv_N20"],
         '^--',
         label="N=20",
         linewidth=2,
         markersize=6)
plt.plot(df_cleaned["Bm_N36"],
         df_cleaned["Pcv_N36"],
         'd--',
         label="N=36",
         linewidth=2,
         markersize=6)

# 設定對數尺度
plt.xscale("log")
plt.yscale("log")

# 軸範圍調整
plt.ylim(1, 200)

# 標籤與標題
plt.xlabel(r"$B_{m}$ (mT)", fontsize=12)
plt.ylabel(r"$P_{core}$ (kW/m³)", fontsize=12)
plt.title("Different Turns Compare (Log Scale)", fontsize=14)
plt.legend(fontsize=10)
plt.grid(which="both", linestyle="--", linewidth=0.5)

# 顯示圖表
plt.show()
```
