```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 讀取 CSV 檔案
# file_path = ".\Data\DRM96A_100kHz_100mT_0-10A-1_ALL.csv"
file_path = ".\Data\CH229060_SCOPE_50mT_2A.csv"
# file_path = ".\Data\DMR96A_300kHz_SCOPE_105mT_TRI.csv"
# file_path = ".\Data\CH229060_SCOPE_50mT_14A.csv"

data = pd.read_csv(file_path)

print(data)

# # 確認資料格式
t = data['TIME']
VR = data['VR']
V2 = data['V2']
VL = data['VL']
IR = data['IR']
# Vout = data['Vout']
```

### 定義計算虛擬電壓的函數

```python
# 定義計算虛擬電壓的函數
def calculate_virtual_voltage(k, V2, VL):
    return V2 - (1 / k) * VL
```

### Find K

```python
# 定義 k 值範圍 (0 ~ 1，步長 0.01)
k_values = np.arange(0.19, 0.2, 0.001)

# 繪製所有 k 值對應的虛擬電壓
for k in k_values:
    # 重新繪製所有 k 值對應的虛擬電壓
    plt.figure(figsize=(10, 6))
    # 繪製原始 VR 波形
    plt.plot(t, VR, label='VR', linewidth=2, color='black')
    Vout = calculate_virtual_voltage(k, V2, VL)
    plt.plot(t, Vout, alpha=0.3, label=f'k={k:.3f}')
    plt.title("Virtual Voltage for Different k Values")
    plt.xlabel("Time (s)")
    plt.ylabel("Voltage (V)")
    plt.grid(True)
    plt.legend(loc='upper right', fontsize='small', ncol=3)
    plt.show()
```
