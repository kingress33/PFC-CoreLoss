```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 讀取 CSV 檔案
# file_path = ".\Data\DRM96A_100kHz_100mT_0-10A-1_ALL.csv"
# file_path = ".\Data\CH229060_SCOPE_50mT_2A.csv"
file_path = ".\Data\DMR96A_300kHz_SCOPE_105mT_TRI.csv"
# file_path = ".\Data\CH229060_SCOPE_50mT_14A.csv"

data = pd.read_csv(file_path)

print(data)

# # 確認資料格式
t = data['TIME']
VR = data['VR']
V2 = data['V2']
VL = data['VL']
IR = data['IR']
Vout = data['Vout']
```

```python
# 繪製SIN用
plt.figure(figsize=(8, 6))
plt.plot(t, VR, label='VR', linewidth=2, color='black')
# plt.plot(t, V2, label='V2')
plt.plot(t, Vout, alpha=0.5, label='V2-(1/k)*VL')
plt.title("Compare w/ and w/o CF ")
plt.xlabel("time (s)")
plt.ylabel("V (V)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#TRI用
Vout_pp = data['Vout_Vpp']
VL_out = data['VL_out']
VL_outpp = data['VL_outpp']

# 繪製TRI用
plt.figure(figsize=(8, 6))
plt.plot(t, VR, label='VR')
plt.plot(t, V2, label='V2')
plt.plot(t, Vout, label='V2-(1/k)*VL(Deskew)')
plt.plot(t, Vout_pp, label='V2-(1/k)*VL')
plt.title("Compare w/ and w/o CF ")
plt.xlabel("time (s)")
plt.ylabel("V (V)")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(t, VL_out, label='(1/k)*VL(Deskew)')
plt.plot(t, VL_outpp, label='(1/k)*VL(Vpp)')
plt.plot(t, V2, label='V2')
plt.title("Compare w/ and w/o CF ")
plt.xlabel("time (s)")
plt.ylabel("V (V)")
plt.grid(True)
plt.legend()
plt.show()
```
