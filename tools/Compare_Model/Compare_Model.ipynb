```python
import numpy as np
import matplotlib.pyplot as plt

# 定義雷達圖的標籤
labels = [
    "Interpretability", "Speed", "Model Complexity", "Data Requirement",
    "Accuracy"
]
num_vars = len(labels)

# 各類模型的數據
data_driven = [1.8, 5, 1, 1, 4]  # 數據驅動模型
physics_driven = [5, 2, 5, 5, 3]  # 物理驅動模型
hybrid = [4, 4, 3, 3, 5]  # 混合模型


# 轉換數據格式以適應雷達圖
def preprocess_data(data):
    return np.array(data + [data[0]])  # 使雷達圖封閉


angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # 封閉雷達圖

# 創建雷達圖
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

# 設置網格線和標籤
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=14, fontweight="bold")
ax.set_yticklabels(["Poor", "", "Good", "", "Best"],
                   fontsize=12,
                   fontweight="bold")
ax.set_ylim(0, 5)  # 設定範圍 0-5
ax.yaxis.grid(True, linestyle="--", linewidth=0.7, alpha=0.7)

# 繪製雷達圖
ax.plot(angles,
        preprocess_data(data_driven),
        color="deepskyblue",
        linewidth=2.5,
        alpha=0.6,
        marker="o",
        label="Data-driven")
# ax.fill(angles, preprocess_data(data_driven), color="deepskyblue", alpha=0.15)

ax.plot(angles,
        preprocess_data(physics_driven),
        color="mediumseagreen",
        linewidth=2.5,
        alpha=0.6,
        marker="o",
        label="Physics-driven")
# ax.fill(angles, preprocess_data(physics_driven), color="mediumseagreen", alpha=0.15)

ax.plot(angles,
        preprocess_data(hybrid),
        color="navy",
        linewidth=2.5,
        alpha=0.6,
        marker="o",
        label="Hybrid")
# ax.fill(angles, preprocess_data(hybrid), color="navy", alpha=0.15)

# 添加標題與圖例
plt.legend(loc="upper center",
           bbox_to_anchor=(0.5, 1.15),
           ncol=3,
           fontsize=12,
           frameon=False)
# plt.title("Radar charts for three categories of dynamic hysteresis models", fontsize=14, fontweight="bold", pad=20)

# 顯示圖表
plt.show()
```
