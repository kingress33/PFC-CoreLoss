```python
import matplotlib.pyplot as plt
import numpy as np

# Define the categories and the values for each model
categories = ['Interpretability', 'Speed', 'Model Complexity', 'Data Requirement', 'Accuracy']
categories = [*categories, categories[0]]  # Repeat the first category to close the radar chart

# Values for each model
data_driven = [1.8, 5, 1, 1, 4]
physics_driven = [5, 2, 5, 5, 3]
hybrid = [4, 4, 3, 3, 5]

# Close the circle for each model
data_driven += data_driven[:1]
physics_driven += physics_driven[:1]
hybrid += hybrid[:1]

# Create the angles for the radar chart
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=True)

# Initialize the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot each model
ax.plot(angles, data_driven, label='Data-driven', color='skyblue', linewidth=2, linestyle='solid')
ax.plot(angles, physics_driven, label='Physics-driven', color='green', linewidth=2, linestyle='solid')
ax.plot(angles, hybrid, label='Hybrid', color='navy', linewidth=2, linestyle='solid')

# Add labels and title
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_yticklabels(['Poor', ' ', 'Good', '', 'Best'])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories[:-1], fontsize=10, weight='bold')

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Display the chart
plt.show()
```
